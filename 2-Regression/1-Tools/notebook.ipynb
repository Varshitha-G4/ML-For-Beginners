```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model, model_selection
```

##### The below command splits the dataset into features (X) and target values (y) => using X, y means you're unpacking the tuples directly into two seperate variables

```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
print(y[0])
```

##### We wanna predict if they have diabetes only using bmi (i.e, 3rd col -- index = 2). Hence we are reducing from multiple cols to only the 3rd col

```python
X = X[:, 2]
```

##### reshape is for keeping all the 442 points (-1 does this) and 1 => 1 column. This reshapes the previously changed 1D array back to 2D (since scikit learn expects 2D inputs)

```python
X = X.reshape((-1, 1))
print(X.shape)
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color="black")
plt.plot(X_test, y_pred, color="red")
plt.xlabel("BMI")
plt.ylabel("Diabetes or not")
plt.show
```
